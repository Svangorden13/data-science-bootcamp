- Question 3 (Multiply all the numbers in a list):
    - Write a function to multiply all the numbers in a list
    - ```py 
        nums = [-7, 5, 6, 120, 30]
      ```
    - The result should look like this:<br>
      ```py
      -756000
      ```

````python
def multiply(nums):
    """This function is to return the multiplication result."""
    m = 1
    for num in nums:
        m *= num
    return m
````

- Question 4 (Check uppercase and lowercase quantity):
    - Write a function to calculate how many uppercase and lowercase letters in a given document.
    - Test your function with the following documents:
        ```py
        doc = ['I love machine learning!',
                'scikit-learn is a free software machine learning library for the Python programming language.',
                'DBSCAN - Density-Based Spatial Clustering of Applications with Noise.'
                ]
        ```
    - The result should look like this:
    
        ```py
        I love machine learning!: {'upper_case': 1, 'lower_case': 19}
        scikit-learn is a free software machine learning library for the Python programming language.: {'upper_case': 1, 'lower_case': 78}
        DBSCAN - Density-Based Spatial Clustering of Applications with Noise.: {'upper_case': 12, 'lower_case': 46}
        ```
    

````python
def check_lower_upper_case(s):
    """This function is to calculate the quantity of lower and upper case letters in a given document."""
    
    d = {"upper_case": 0, "lower_case": 0}
    
    for i in s:
        if i.islower():
            d['lower_case'] += 1
        elif i.isupper():
            d['upper_case'] += 1
        else:
            pass
    
    return d
````

````python
doc = ['I love machine learning!',
       'scikit-learn is a free software machine learning library for the Python programming language.',
       'DBSCAN - Density-Based Spatial Clustering of Applications with Noise.'
       ]
````

````python
for i in doc:
    print(f"{i}: {check_lower_upper_case(i)}")
````

- Question 5 (Find unique elements in a list):
    - Write a function to return a list with unique elements
    - Test your function with the following lists:
        ```py
        nums1 = [1, 1, 1, 2, 2, 3, 4, 7, 7, 7, 10]
        nums2 = [1, 2, 3, 4]
        ```
    - The result should look like this:
        ```py
        The unique elements in the list [1, 1, 1, 2, 2, 3, 4, 7, 7, 7, 10]: [1, 2, 3, 4, 7, 10]
        The unique elements in the list [1, 2, 3, 4]: [1, 2, 3, 4]
        ```

````python
def find_unique_elements(nums):
    """This funciton is to find unique elements in a given list."""
    
    unique_lst = []
    
    for num in nums:
        if num not in unique_lst:
            unique_lst.append(num)
    
    return unique_lst
            
````

````python
nums = [
    [1, 1, 1, 2, 2, 3, 4, 7, 7, 7, 10],
    [1, 2, 3, 4]
]
````

````python
for i in nums:
    print(f"The unique elements in the list {i}: {find_unique_elements(i)}")
````

- Question 7 (Sort a string):
    - Write a function with an input parameter accepting hyphen-separated sequence of words <br>
      and return this string in a hyphen-separated sequence of words after sorting them alphabetically.
    - Test your function with the string, 
      ```py
      "mango-strawberry-apple-guava"
      ```
    - The result should look like this:
      ```py
      'apple-guava-mango-strawberry'
      ```

````python
def sort_strings(s):
    """This function is to sort a hyphen-separated sequence of words alphabetically."""
    
    return  "-".join(sorted(s.split("-")))
````

````python
s = 'mango-strawberry-apple-guava'

sort_strings(s)
````

- Question 9 (Mean Squared Error):
    - In statistics, mean squred error (MSE) is the average squared differences between the true and prediction value.<br>
      Can you write a function to take 2 input paramters, `y` and `y_pred` to  calcuate the MSE of the following 2 lists?<br>
      ```py
      y = [1, 2, 3] # true value
      y_pred = [1.1, 1.8, 3.5] # prediction value
      ```
    &emsp;
    - Hint: How to calculate the MSE?
      - step 1: get the squared differences of each element in `y` and `y_pred`
        - ex: get the squared differences of the first element in both list
          ```py
          (1 - 1.1) ** 2 = 0.01
          ```
        - ex: get the squared differences of the second element in both list
          ```py
          (2 - 1.8) ** 2 = 0.04
          ```
        - ex: get the squared differences of the third element in both list
          ```py
          (3 - 3.5) ** 2 = 0.25
          ```
      - step 2: append each squared differences into a new list
      - step 3: sum this new list and divided by the length of this new list
    - Note: round up the result to the 2nd decimal place
    - The result should look like this:<br>
      ```py
        0.1
      ```

````python
y = [1, 2, 3]
y_pred = [1.1, 1.8, 3.5]
````

````python
def mean_sq_error(y, y_hat):
    """This function is to calculate the MSE of true and prediction values."""
    
    sq_error = []
    for i in range(len(y)):
        error = (y[i] - y_hat[i]) ** 2
        sq_error.append(error)
    
    mse = sum(sq_error)/len(sq_error)
    
    return round(mse, 2)
        
````

````python
mean_sq_error(y, y_pred)
````

- Question 10 (Standardize):
    - Write a function called `standardize` and it will take 3 input parameters, `x`, `x_center`, `scale_size`.<br>
      This function will return the difference between `x` and `x_center`, divided by `scale_size`.<br>
      Call this function with ***keyword arguments*** and try to switch the arguments order with each call.<br>
<br>
    - The result should look like this:<br>
      ```py
      standardize(x=8.2, x_center=13.8, scale_size=4.83): -1.1594202898550727
      standardize(scale_size=24.63, x=2.89, x_center=-72.813): 3.073609419407227
      ```

````python
def standardize(x, x_center, scale_size):
  """This function is to standardize the number."""
  
  return (x-x_center)/scale_size

print(f"standardize(x=8.2, x_center=13.8, scale_size=4.83): {standardize(x=8.2, x_center=13.8, scale_size=4.83)}")
print(f"standardize(scale_size=24.63, x=2.89, x_center=-72.813): {standardize(scale_size=24.63, x=2.89, x_center=-72.813)}")
````
