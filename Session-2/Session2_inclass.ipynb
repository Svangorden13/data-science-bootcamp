# Intro to Python

Ask for the age of the person and then:

*   If the person is less than 2 years old, print a message that the person is a baby.
*   If the person is at least 2 years old but less than 4, print a message that the person is a toddler.
*   If the person is at least 4 years old but less than 13, print a message that the person is a kid.
*   If the person is at least 13 years old but less than 20, print a message that the person is a teenager.
*   If the person is at least 20 years old but less than 65, print a message that the person is an adult.







```python
age = int(input("Age :"))

if age < 2:
  print("Child")
elif age < 4:
  print("Toddler")
elif age < 13:
  print("kid")
elif age < 20:
  print("teenager")
elif age < 65:
  print("Adult")
```

```python
grade = float(input("Enter a score: "))
if grade < 60:
    print("Letter grade is " + "F")
elif grade < 70:
    print("Letter grade is " + "D")
elif grade < 80:
    print("Letter grade is " + "C")
elif grade < 90:
    print("Letter grade is " + "B")
else:
    print("Letter grade is " + "A")
```

Exercise :

Write a loop that prompts the user to enter a series of pizza toppings until they enter a 'quit' value. As they enter each topping, print a message saying you’ll add that topping to their pizza.

*   Use a break statement to exit the loop when the user enters a 'quit' value.
*   Maximum of 4 toppings allowed



```python

while True:
  topping = input("Add topping for your pizza : ")
  if topping != 'quit':
    print("You have added {}".format(topping))
  else:
    print('Thank you for ordering')
    break

```

```python
i= 1
while i < 4:
  i += 1
  topping = input("Add topping for your pizza : ")
  if topping != 'quit':
    print("You have added {}".format(topping))
  else:
    print('Thank you for ordering')
    break
else:
  print('You have reached the maximum of 3 toppings allowed')
```
