```python
import numpy as np
```

```python
def cov_a(list1,list2):
  mean_list1 = np.mean(list1)
  mean_list2 = np.mean(list2)
  num = 0
  for i in range(0,len(list1)):
    for j in range(0,len(list2)):
      if i == j:
        num += (list1[i] - mean_list1) * (list2[i] - mean_list2)
      else:
        pass
  den = len(list1) - 1
  return num/den
```

```python
cov_a([18,25,40,50],[125,120,100,70])
```

# Balancing multi-class


```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
```

```python
data = titanic[['Age','Fare','Embarked']]
data.Age = data.Age.fillna(0)
data.Embarked = data.Embarked.fillna('UN')
data
```

```python
data.Embarked.value_counts()
```

```python
X, y = data.iloc[:, :-1], data.iloc[:, -1]
X.shape, y.shape
```

```python
y
```

```python
# label encode the target variable
y = LabelEncoder().fit_transform(y)
y
```

```python
oversample = SMOTE(k_neighbors=1, )
X, y = oversample.fit_resample(X,y)
```
